```python
# 1. fungsi untuk membaca data
def read_data (emisi_file,kebakaran_file,geo_file) :
  emisi = pd.read_csv(emisi_file)
  kebakaran = pd.read_csv(kebakaran_file)
  geo = pd.read_csv(geo_file)
  return emisi, kebakaran, geo
```

```python
# 2 fungsi untuk menhapus baris dengan nilai null
def hapus_null(df_input):
  df_input.dropna(axis="columns", how=all, inplace=True)
  df_input = df_input.dropna(axis="index", thresh=2, inplace=False)
  return df_input
```

```python
# 3 fungsi mengisi missing value
def isi_missing(df_input):
  df_input['STATUS'].fillna("Member State", inplace=True)
  return df_input
```

```python
# 4 Fungsi untung menggabungkan 3 data
def gabung_data(df1,df2,df3):
  df_gg1=pd.merge(df1,df2,on="Provinsi")
  df_gg=pd.merge(df3,df_gg1,on="Provinsi")
  return df_gg
```

```python
# 5 Fungsi rename kolom
def rename_kolom (df_input) :
  df_input.rename(columns={"2018_x":"Jumlah_Emisi_2018","2019_x":"Jumlah_Emisi_2019","2020_x":"Jumlah_Emisi_2020","2021_x":"Jumlah_Emisi_2021","2022_x":"Jumlah_Emisi_2022", "2018_y":"jumlah_Kebakaran_Hutan_2018", "2019_y":"jumlah_Kebakaran_Hutan_2019", "2020_y":"jumlah_Kebakaran_Hutan_2020", "2021_y":"jumlah_Kebakaran_Hutan_2021", "2022_y":"jumlah_Kebakaran_Hutan_2022"},inplace=True)
  return df_input
```

```python
# 1 pipeline
def pipeline_read(emisi_file,kebakaran_file,geo_file):
  df_emisi=pd.DataFrame()
  df_kebakaran=pd.DataFrame()
  df_geo=pd.DataFrame
  try:
    df_emisi=read_data(emisi_file,kebakaran_file,geo_file)[0]
    df_kebakaran=read_data(emisi_file,kebakaran_file,geo_file)[1]
    df_geo=read_data(emisi_file,kebakaran_file,geo_file)[2]
  except Exception as e:
    logging.info("Baca data gagal")
  return df_emisi,df_kebakaran,df_geo
```

```python
# pipeline hapus baris null
def pipeline_hapus(df_bersih,df_bersih2,df_bersih3):
  try:
    df_bersih=hapus_null(df_bersih)
    df_bersih2=hapus_null(df_bersih2)
    df_bersih3=hapus_null(df_bersih3)

  except Exception as e:
    logging.info("Hapus baris null gagal")
  return df_bersih,df_bersih2,df_bersih3
```

```python
#pipeline missing value
def pipeline_isi(df_bersih,df_bersih2,df_bersih3):
  try:
    df_berish=isi_missing(df_bersih)
    df_bersih2=isi_missing(df_bersih2)
    df_bersih3=isi_missing(df_bersih3)

  except:
    logging.info("Isi Missing Value gagal")
    return df_bersih,df_bersih2,df_bersih3
```

```python
#pipeline merge data
def pipeline_gabung (df_isi,df_isi2,df_isi3):
  try:
    df_gabung=gabung_data(df_isi,df_isi2,df_isi3)
  except:
    logging.info("Gabung data gagal")
  return df_gabung
```

```python
#pipeline rename
def pipeline_rename(df_gabung):
  try:
    df_gg=rename_kolom(df_gabung)
  except:
    logging.info("Rename gagal")
  return df_gg
```

```python
x,y,z=pipeline_read(data_Emisi,data_luaskebakaran,data_geo)
xbersih,ybersih,zbersih=pipeline_hapus(x,y,z)
xisi,yisi,zisi=pipeline_isi(xbersih,ybersih,zbersih)
xgabung=pipeline_gabung(xisi,yisi,zisi)
xgg=pipeline_rename(xgabung)
xgg.head()
```
